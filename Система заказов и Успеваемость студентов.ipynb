```python
menu = {}
orders = {}

def add_dish(name, price, category, cook_time):
    menu[name] = {"цена": price, "категория": category, "время": cook_time, "заказов": 0}

def create_order(table, dishes):
    orders[table] = {"блюда": dishes, "статус": "готовится"}
    for dish in dishes:
        if dish in menu:
            menu[dish]["заказов"] += 1

def calculate_order(table):
    total = sum(menu[d]["цена"] for d in orders[table]["блюда"] if d in menu)
    return total

def popular_dishes():
    sorted_dishes = sorted(menu.items(), key=lambda x: x[1]["заказов"], reverse=True)
    return [(name, info["заказов"]) for name, info in sorted_dishes[:3]]

def revenue_by_category():
    revenue = {}
    for table, order in orders.items():
        for dish in order["блюда"]:
            if dish in menu:
                cat = menu[dish]["категория"]
                revenue[cat] = revenue.get(cat, 0) + menu[dish]["цена"]
    return revenue

# Тест
add_dish("Пицца", 500, "основное", 20)
add_dish("Кофе", 150, "напитки", 5)
add_dish("Суп", 300, "основное", 15)
add_dish("Чай", 100, "напитки", 3)

create_order(1, ["Пицца", "Кофе"])
create_order(2, ["Суп", "Чай", "Кофе"])

print("Стоимость стола 1:", calculate_order(1), "руб.")
print("Популярные блюда:", popular_dishes())
print("Выручка по категориям:", revenue_by_category())
```

```python
students = {}
grades = []

def add_student(student_id, name, group, contact):
    students[student_id] = {"фио": name, "группа": group, "контакт": contact}

def add_grade(student_id, subject, grade, date):
    grades.append({"id": student_id, "предмет": subject, "оценка": grade, "дата": date})

def average_grade(student_id):
    student_grades = [g["оценка"] for g in grades if g["id"] == student_id]
    return sum(student_grades) / len(student_grades) if student_grades else 0

def low_performers(threshold):
    result = []
    for sid in students:
        if average_grade(sid) < threshold:
            result.append((students[sid]["фио"], average_grade(sid)))
    return result

def rating():
    ranked = [(students[sid]["фио"], average_grade(sid)) for sid in students]
    return sorted(ranked, key=lambda x: x[1], reverse=True)

# Тест
add_student(1, "Александр Черненко", "ИC-33", "alexandr@mail.ru")
add_student(2, "Матвей Чернявский", "ИC-33", "matvey@mail.ru")
add_student(3, "Иван Иванов", "ИC-33", "ivan@mail.ru")

add_grade(1, "Python", 5, "2024-01-15")
add_grade(1, "Математика", 4, "2024-01-16")
add_grade(2, "Python", 5, "2024-01-15")
add_grade(2, "Математика", 5, "2024-01-16")
add_grade(3, "Python", 3, "2024-01-15")
add_grade(3, "Математика", 3, "2024-01-16")

print("Средний балл Александра:", average_grade(1))
print("Слабые студенты (ниже 4):", low_performers(4))
print("Рейтинг:", rating())
```
